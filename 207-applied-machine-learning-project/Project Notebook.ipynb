# Prepping data for EDA

### Combining the Centroids files

```python
import pandas as pd


# State Labeled Centroid Files
ga = pd.read_csv("labeled_GA_centroids.csv")
sc = pd.read_csv("labeled_SC_centroids.csv")
ky = pd.read_csv("labeled_Kentucky_centroids.csv")
tn = pd.read_csv("labeled_Tennessee_centroids.csv")
ar = pd.read_csv("labeled_Arkansas_centroids.csv")
al = pd.read_csv("labeled_Alabama_centroids.csv")
la = pd.read_csv("labeled_Louisiana_centroids.csv")
ms = pd.read_csv("labeled_Miss_centroids.csv")

# Add state column 
ga["state"] = "GA"
sc["state"] = "SC"
ky["state"] = "Kentucky"
tn["state"] = "Tennessee"
ar["state"] = "Arkansas"
al["state"] = "Alabama"
la["state"] = "Louisiana"
ms["state"] = "Mississippi"

# Combine
all_data = pd.concat([ga, sc, ky, tn, ar, al, la, ms], ignore_index=True)
all_data.to_csv("regional_combined_centroids.csv", index=False)
```
